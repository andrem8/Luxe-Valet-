```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
"""Hypothesis: Distracted drivers are more likely to incur driving infractions. To test hypothesis graph the correlation of phone
usage to driving infractions and compare with null correlation."""
```

```python
Luxe_Data = pd.read_csv('Analysis Exercise - Safety2.csv')
```

```python
#Convert Data to matrix array, then remove zeros from array, convert to list
Phone_Usage = Luxe_Data['Avg # of Phone Usages per 100 mile (secs)'].as_matrix()
Infractions = Luxe_Data['Infractions per 100 Miles'].as_matrix()
PhU = np.trim_zeros(Phone_Usage)
Inf = np.trim_zeros(Infractions)
PhU.tolist()
Inf.tolist()
Inf[:3]
```

```python
# setup graph of best fit line
x = Inf
y = PhU

# Find the slope and intercept of the best fit line
slope, intercept = np.polyfit(x, y, 1)

# Create a list of values in the best fit line
abline_values = [slope * i + intercept for i in x]

# Plot the best fit line over the actual values
plt.plot(x, y, '--')
plt.plot(x, abline_values, 'b')
plt.title(slope)
plt.show()

```

```python
#this graph shows the correlation of #phoneusage to #driving infractions
```

```python
Luxe_Data1 = pd.read_csv('Analysis Exercise - Safety1.csv')
User_ID = Luxe_Data1['User ID'].as_matrix()
User = np.trim_zeros(User_ID)
User.tolist()
User[:3]
```

```python
#this graph shows the correlation of user ids to driving infractions. There is no expected correlation.
```

```python
x = User
y = Inf
slope, intercept = np.polyfit(x, y, 1)
abline_values = [slope * i + intercept for i in x]
plt.plot(x, y, '--')
plt.plot(x, abline_values, 'b')
plt.title(slope)
plt.show()
```

```python
"""From the examination of the first graph we can see that there is a strong correlation between
phone usage and driving infractions.  As expected, there is no correlation between user ids and
driving infractions. To improve valet service it is recommended to develop KPIs monitoring
phone usage and to supply handsfree technology to valets."""
```
